```python
from huggingface_hub import InferenceClient
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()

HF_TOKEN = user_secrets.get_secret('HUGGING_FACEE_API_KEY')
client = InferenceClient(api_key=HF_TOKEN)
```

```python
modelo = "openai/gpt-oss-120b"
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from huggingface_hub import InferenceClient

HF_TOKEN = user_secrets.get_secret("HUGGING_FACEE_API_KEY")
client = InferenceClient(api_key=HF_TOKEN)


texto_input = widgets.Textarea(
    placeholder="Escolha a persona: operador comercial, síndico, morador ou técnico...",
    description="Pergunta:",
    layout={"width": "100%", "height": "100px"}
)


botao_enviar = widgets.Button(
    description="Gerar Resposta",
    button_style="success",
    icon="paper-plane"
)


output_area = widgets.Output(
    layout={"border": "1px solid #ccc", "padding": "10px", "margin": "10px 0"}
)


def ao_clicar_no_botao(b):
    with output_area:
        clear_output()
        print("Pensando...")

        try:
            resposta = client.chat_completion(
                model=modelo,
                messages=[
                    {
                        "role": "system",
                        "content": "Você responde resolvendo o seguinte problema de acordo com a escolha do usuário entre (operador comercial, síndico, morador ou técnico) NENHUMA OUTRA PERSONA: A ausência de mecanismos integrados em eletropostos comerciais para orquestrar potência, registrar ciclos, faturar e comunicar, ou seja, a integração de bombas elétricas em postos de gasolina para que seja viável o carregamento do carro em um tempo curto e eficiente. Utilizando técnicas do chatbot, incluindo tecnologias escolhidas, fluxo de funcionamento e modelo de teste."
                    },
                    {
                        "role": "user",
                        "content": texto_input.value
                    }
                ],
                temperature=0.75,
                top_p=0.9,
            )

            clear_output()
            print(resposta.choices[0].message.content)

        except Exception as e:
            clear_output()
            print(f"Erro na chamada: {e}")


botao_enviar.on_click(ao_clicar_no_botao)


display(texto_input, botao_enviar, output_area)
```
